```python
df = pd.read_csv("parsing_results/shows_fixed.csv")

display(df.head())
```

```python
display(df.iloc[1, 10])
display(len(df.iloc[1, 10]))
print(df.loc[1, ["whats_the_point"]])
```

```python
df.dropna(subset=["start_date"], inplace=True)
```

```python
df["whats_the_point"] = df["whats_the_point"].fillna("")
```

```python
df["description"] = df["description"] + "\n" + df["whats_the_point"]
df = df.drop("whats_the_point", axis=1)
```

```python
display(df.iloc[13, [10, 11]])
print(df.iloc[13, 10])
print(df.iloc[13, 11])
```

```python
df = df.drop("Unnamed: 0", axis=1)
```

```python
df.isna().sum()
```

```python
df[df["start_date"].isna()]
```

```python
df.shape
```

```python
df["ongoing"] = df["end_date"].apply(lambda x: 1 if pd.isna(x) else 0)
df[df["end_date"].isna()]
```

```python
df["episodes"] = df["episodes"].fillna(0)
df[df["episodes"].isna()]
```

```python
current_year = int(pd.Timestamp.now().year)
df["end_date"] = df["end_date"].fillna(current_year)
df[df["end_date"].isna()]
```

```python
df[["start_date", "end_date", "episode_duration_minutes", "episodes"]] = df[
    ["start_date", "end_date", "episode_duration_minutes", "episodes"]
].astype(int)
df.head(3)
```

```python
base_columns = df.columns.to_list()
base_columns
```

```python
df[df["genres"] == "Аниме,"]
```

```python
genres_dummies = df["genres"].str.get_dummies(" , ")

# Объединяем с исходным DataFrame
df = pd.concat([df, genres_dummies], axis=1)
```

```python
for c in df.columns.to_list():
    df.rename(columns={c: c.strip()})
```
